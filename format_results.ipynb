```python
def to_label(x, n_classes=2):
    if isinstance(x, str):
        x = x.replace(",", ".")
        x = float(x)

    if n_classes == 2:
        if x >= 2.5:
            return "olumlu"
        else:
            return "olumsuz"
    elif n_classes == 3:
        if x >= 3:
            return "olumlu"
        elif x >= 2 and x < 3:
            return "nötr"
        else:
            return "olumsuz"
    
```

```python
def load_dataset(dataset):
    dataset = dataset.copy()

    test_size = dataset.pop("test_size")
    return (
        Dataset(**dataset)
        .load_dataset()
        .map_labels()
        .preprocess("text")
        .preprocess("label")
        .train_test_split(test_size=test_size, random_state=7)
    )
```
